### Cleaning the Data

```python
## Loading Dataset

data = pd.read_csv('https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv')

## Removing Duplicate Rows

data.drop_duplicates(inplace=True)

## Replacing White Space in Column Names

def clean_column_names(df):
    def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()

    df.columns = [clean_name(col) for col in df.columns]
    return df

data = clean_column_names(data)


## Dropping Rows with Missing Values

data.replace(to_replace= 'missing', value= None, inplace= True)

data.dropna(inplace= True)

data

### There were no values with None, only values with the string 'missing'.
```
